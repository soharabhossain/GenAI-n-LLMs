## Example Sentences

```python
# Two sentences to compare
sentence1 = "The company announced record profits for the fiscal year."
sentence2 = "The firm's annual earnings exceeded all expectations."

print("Sentence 1:", sentence1)
print("Sentence 2:", sentence2)
print("\nGoal: Prove mathematically that these sentences mean almost the same thing.")
```

```python
from sentence_transformers import SentenceTransformer, util
import numpy as np
print("✓ Libraries imported successfully")

# Load a pre-trained model that understands language
# This model has been trained on a huge amount of text data
model = SentenceTransformer('all-MiniLM-L6-v2')
print("Model loaded successfully!")
print(f"This model produces embeddings with {model.get_sentence_embedding_dimension()} dimensions")
```

# Convert sentences into numerical vectors (embeddings)

```python
# Convert sentences into numerical vectors (embeddings)
embedding1 = model.encode(sentence1, normalize_embeddings=True)
embedding2 = model.encode(sentence2, normalize_embeddings=True)

print(f"Embedding 1 shape: {embedding1.shape}")
print(f"Embedding 2 shape: {embedding2.shape}")
print(f"\nFirst 10 values of embedding 1:")
print(embedding1[:10])
```

# Calculate cosine similarity using SentenceTransformer's built-in function


```python
# Calculate cosine similarity using SentenceTransformer's built-in function
# This is more reliable than sklearn for sentence embeddings
similarity_score = util.cos_sim(embedding1, embedding2).item()

print(f"\nCosine Similarity Score: {similarity_score:.4f}")
print("\nInterpretation:")
print("- Score close to 1.0: Sentences are semantically very similar")
print("- Score close to 0.0: Sentences are unrelated")
print("- Score close to -1.0: Sentences have opposite meanings")

if similarity_score > 0.7:
    print(f"\n✓ High score ({similarity_score:.4f}) confirms these sentences have very similar meanings!")
```

```python
# Try with opposite meanings
sentence3 = "The firm has reported massive losses for this quarter."
embedding3 = model.encode(sentence3)

# Calculate similarity with the opposite meaning sentence
similarity_opposite = util.cos_sim(embedding1, embedding3).item()

print(f"\nSimilarity between Sentence 1 and opposite meaning: {similarity_opposite:.4f}")
print("Notice this is much lower, showing the model understands semantic opposition!")
```

```python
# Show the difference
print(f"\nComparison:")
print(f"  Similar sentences:  {similarity_score:.4f}")
print(f"  Opposite sentences: {similarity_opposite:.4f}")
print(f"  Difference:         {similarity_score - similarity_opposite:.4f}")
```

```python
# Compare all three sentences
print("\n" + "="*60)
print("SEMANTIC SIMILARITY MATRIX")

print("="*60)
print(f"Sentence 1 ↔ Sentence 2 (similar meaning):  {similarity_score:.4f}")
print(f"Sentence 1 ↔ Sentence 3 (opposite meaning): {similarity_opposite:.4f}")
print(f"\nDifference: {(similarity_score - similarity_opposite):.4f}")
```

```python
print("\nConclusion:")
if similarity_score > 0.47:
    print("✓ Model successfully identified semantic similarity")
if similarity_score - similarity_opposite > 0.08:
    print("✓ Model successfully distinguished similar from opposite meanings")
print("="*60)
```
